## Exercise 1 - rlanders

Load the dataset data/rlanders.csv. Use the column Money.  
We assume that the values from this sample are normally distributed around a population mean $\mu$ that is unknown, but that the standard deviation of the population is known, i.e. $\sigma = 98$.


```python
df = pd.read_csv('https://raw.githubusercontent.com/HoGentTIN/dsai-en-labs/main/data/rlanders.csv').set_index(['ID'])
# Show the first few records of the Data Frame
df.head()
```

Calculate a 99% confidence interval for the population mean (Answer: [484.191, 516.121]).

```python
# gemiddelde van sample
m = df['Money'].mean()

# op basis van gemiddelde berekenen we confidence interval
# we kennen de std / sigma maar popluatie niet

sigma = 98
# alpha = 0.01
n = df['Money'].count()


stats.norm.interval(0.99, loc=m, scale=sigma / n**0.5)
```

Calculate a 95% confidence interval for the population mean. (Answer =  [488.008, 512.304])

```python
stats.norm.interval(0.95, loc=m, scale=sigma / n**0.5)
```

Suppose σ is unknown, calculate a 95% confidence interval for the population mean. (Answer = [487.319, 512.993])

```python
# als sigma unknown -> Student t verdeling

sigma = df['Money'].std()
alpha = 0.05

stats.t.interval(1-alpha, df=n-1, loc=m, scale=sigma/n**0.5)
```

Finally, suppose the sample consists of only the first 25 observations in this dataset. Calculate a 95% confidence interval for this situation.  
(Answer = [450.291,536.669])

```python
# als n < 30

n = 25
mu = df['Money'][:25].mean()
sigma = df['Money'][:25].std()
alpha = 0.05

stats.t.interval(1-alpha, df = n - 1, loc=mu, scale=sigma/np.sqrt(n))
```

## Exercise 2

Intelligentie is normaal verdeeld met een gemiddelde van 100 en een standaardafwijking van 15.

1. Om toegelaten te worden tot Mensa moet je tot de 2% slimsten van de bevolking behoren. Met welk IQ stemt dit overeen?

2. Je doet op het internet een IQ test en er verschijnt dat jouw IQ ongeveer 120 is. Welk percentage van de bevolking heeft een IQ van 120 of meer?

3. Mensen met een IQ tussen de 70 en 85 zijn zwakbegaafd. Over hoeveel procent van de bevolking gaat het?

```python
mu = 100
sigma = 15

# 1. Om toegelaten te worden tot Mensa moet je tot de 2% slimsten van de bevolking behoren.
# Met welk IQ stemt dit overeen?

stats.norm.isf(0.02, loc=mu, scale=sigma)

# 2. Je doet op het internet een IQ test en er verschijnt dat jouw IQ ongeveer 120 is.
# Welk percentage van de bevolking heeft een IQ van 120 of meer?

stats.norm.sf(120, loc=mu, scale=sigma)


# 3. Mensen met een IQ tussen de 70 en 85 zijn zwakbegaafd.
# Over hoeveel procent van de bevolking gaat het?

stats.norm.cdf(85, loc=mu, scale=sigma) - stats.norm.cdf(70, loc=mu, scale=sigma)
```
